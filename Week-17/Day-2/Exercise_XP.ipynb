```python
temperature = [30, 31, 29, 32, 33, 34, 28, 27, 25, 35]
humidity = [80, 75, 70, 85, 90, 80, 75, 60, 65, 70]
wind_speed = [5, 6, 7, 8, 9, 4, 10, 5, 6, 7]

date_rng = pd.date_range(start = '2021-01-01',end = '2021-01-10')

weather_data = pd.DataFrame({'date_range': date_rng, 'temperature': temperature, 'humidity': humidity, 'wind_speed': wind_speed})
weather_data
```

```python
freq = pd.infer_freq(date_rng)
print(f'The freqiency of the dataset is {freq}')
```
