```python
df['Caffeine (mg)'] = pd.to_numeric(df['Caffeine (mg)'], errors='coerce')
```

```python
#11
df['caffeine_to_calories'] = df['Caffeine (mg)'] / df['Calories']
df.head()

```

```python
#12

df.groupby('Beverage_category')['caffeine_to_calories'].mean().sort_values(ascending=False)
```

```python
#13
plt.barh(df['Beverage_category'],df['caffeine_to_calories'])
plt.show()
```

```python
#14
#Question = What is the relationship between Calories and Cholesterol in drinks

plt.scatter(df['Calories'], df['Cholesterol (mg)'])
plt.xlabel('Calories')
plt.ylabel('Cholesterol')
plt.title('Calories vs Cholesterol')
plt.show()
```
