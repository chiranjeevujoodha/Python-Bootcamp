Exercise 4

```python
x = [1, 2, 3, 4, 5]  # Age
y = [75, 80, 88, 95, 105]  # Height in cm

plt.scatter(x,y)
plt.title('Product Demand vs Price')
plt.xlabel('Price')
plt.ylabel('Demand')
plt.show()

plt.savefig('figure1.png')
```

```python
x = [1, 2, 3, 4, 5]  # Age
y = [75, 80, 88, 95, 105]  # Height in cm

plt.scatter(x,y)
plt.title('Childrens Height vs Age')
plt.xlabel('Age (years)')
plt.ylabel('Height (cm)')
plt.show()

plt.savefig('figure2.png')
```
